```python
import pandas as pd
```

```python
# read in data
movie_df = pd.read_csv("output/movie.csv")
genre_df = pd.read_csv("output/genre.csv")
budget_df = pd.read_csv("output/clean_budget_data.csv")
orig_movie_list = pd.read_csv("output/complete_movie_list_df.csv")
buzz_df = pd.read_csv("output/complete_buzz_data.csv")
```

```python
#drop duplicate titles
orig_movie_list.drop_duplicates(["Title"], keep="first", inplace=True)
```

```python
# create new dfs for tables
new_genre_df = pd.DataFrame({"genre_name": []})
```

```python
# go through genre_df and find all unique genres
for index, row in genre_df.iterrows():
    row_genres = row["genre_name"].split(", ")
    for genre in row_genres:
        if genre not in new_genre_df["genre_name"].values:
            new_genre_row = {'genre_name': genre}
            new_genre_df = new_genre_df.append(new_genre_row, ignore_index=True)
```

```python
new_genre_df.index.name = "genre_id"
new_genre_df
```

```python
budget_df
```

```python
movie_df
```

```python
movie_df["budget"] = "n/a"
movie_df["boxoffice"] = "n/a"
```

```python
for index, row in budget_df.iterrows():
    movie_index = movie_df.index[movie_df["title"] == row["Movie"]].tolist()
    try: 
        
        movie_df.at[movie_df.loc[movie_index[0]]["movie_id"], "budget"] = row["Production Budget"]
        movie_df.at[movie_df.loc[movie_index[0]]["movie_id"], "boxoffice"] = row["Domestic Gross"]
    except IndexError:
        print(f"no such movie {row['Movie']}")
    
```

```python
movie_df.dropna(inplace=True)
```

```python
movie_df.loc[movie_df["budget"] != "n/a"]
```

```python
movie_df = movie_df.loc[movie_df["budget"] != "n/a"]
```

```python
movie_df
```

```python
movie_df["budget"] = pd.to_numeric(movie_df["budget"])
movie_df["boxoffice"] = pd.to_numeric(movie_df["boxoffice"])
```

```python
movie_df
```

```python
movie_genre_junction_df = pd.DataFrame({"movie_id": [], "genre_id": []})
```

```python
for index, row in orig_movie_list.iterrows():
    # check if movie is in new df
    if row["Title"] in movie_df["title"].values:
        # find genre for each movie
        genre_list = row["Genre"].split(", ")
        # add row to junction table for each genre
        for genre in genre_list:
            # print(movie_df["title"] == row["Title"])
            movie_index = movie_df.loc[(movie_df["title"] == row["Title"])].index.tolist()
            movie_id = movie_df.loc[movie_index[0],"movie_id"]
            genre_id = new_genre_df.loc[(new_genre_df["genre_name"] == genre)].index.tolist()
            new_row = {"movie_id": int(movie_id), "genre_id": int(genre_id[0])}
            print(new_row)
            movie_genre_junction_df = movie_genre_junction_df.append(new_row, ignore_index=True)
```

```python
movie_genre_junction_df["genre_id"] = movie_genre_junction_df["genre_id"].astype(int)
movie_genre_junction_df["movie_id"] = movie_genre_junction_df["movie_id"].astype(int)
```

```python
movie_genre_junction_df
```

```python
movie_df["buzz"] = 0
```

```python
for index, row in buzz_df.iterrows():
    # check if movie is in new df
    if row["title"] in movie_df["title"].values:
        # get movie_id
        movie_id = movie_df.loc[(movie_df["title"] == row["title"])].index.tolist()
        # add buzz
        movie_df.loc[movie_id[0], ["buzz"]] = row['buzz']
```
