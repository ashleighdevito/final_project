```python
#T.Bradford
#October 2021

# Dependencies
import requests
from config import api_key
from bs4 import BeautifulSoup as bs
import requests
import numpy as np
import time
import pandas as pd
import csv
import json
```

```python
# Create a Beautiful Soup object
def scrape():
    pages = np.arange(1,501,100)
    years = np.arange(2006,2020,1)
    movies = []
    for year in years:
        #year = requests.get('https://www.the-numbers.com/box-office-records/domestic/all-movies/cumulative/released-in-' + str(year))
        
        for page in pages:
            page = requests.get('https://www.the-numbers.com/box-office-records/domestic/all-movies/cumulative/released-in-' + str(year) + '/' + str(page))
            soup = bs(page.text.encode('latin1').decode('utf-8'), 'html.parser')
            tds = soup.find_all('td')
            #print(tds)

            title = []
            
            # Loop over td elements
            for td in tds:
                # If td element has an anchor...
                if (td.b):
                    # And the anchor has non-blank text...
                    if (td.b.text):
                        # Append the td to the list
                        title.append(td)
            #print(headlines)
            
            for x in range(100):
                movies.append(title[x].text)
            #movies
            #print(movies)
    return(movies)
```

```python
# Create a Beautiful Soup object
def scrape2020_2021():
    pages = np.arange(1,201,100)
    years = np.arange(2020,2022,1)
    movies = []
    for year in years:
        #year = requests.get('https://www.the-numbers.com/box-office-records/domestic/all-movies/cumulative/released-in-' + str(year))
        
        for page in pages:
            page = requests.get('https://www.the-numbers.com/box-office-records/domestic/all-movies/cumulative/released-in-' + str(year) + '/' + str(page))
            soup = bs(page.text.encode('latin1').decode('utf-8'), 'html.parser')
            tds = soup.find_all('td')
            #print(tds)

            title = []
            
            # Loop over td elements
            for td in tds:
                # If td element has an anchor...
                if (td.b):
                    # And the anchor has non-blank text...
                    if (td.b.text):
                        # Append the td to the list
                        title.append(td)
            #print(headlines)
            
            for x in range(100):
                movies.append(title[x].text)
            #movies
            #print(movies)
    return(movies)
```

```python
#API call section 
url = "http://www.omdbapi.com/?apikey=" + api_key + "&t="

#test = ['sing', ' Spider Man 3', 'Black Widow']
#responses = []

responses = {'Title': [], 'Year': [], 'Rated': [], 'Released': [], 'Runtime': [], 'Genre': [], 'Director': [], 'Writer': [], 'Actors': [], 
'Plot': [], 'Language': [], 'Poster': [], 'Metascore': [], 'imdbRating': [], 'BoxOffice': [], 'Production': []}

i = 1
j = 1
for movie in scrape():
    movie_data = requests.get(url + movie).json()
    #responses.append(movie_data)
    try:
        responses['Title'].append(movie_data['Title'])
        responses['Year'].append(movie_data['Year'])
        responses['Rated'].append(movie_data['Rated'])
        responses['Released'].append(movie_data['Released'])
        responses['Runtime'].append(movie_data['Runtime'])
        responses['Genre'].append(movie_data['Genre'])
        responses['Director'].append(movie_data['Director'])
        responses['Writer'].append(movie_data['Writer'])
        responses['Actors'].append(movie_data['Actors'])
        responses['Plot'].append(movie_data['Plot'])
        responses['Language'].append(movie_data['Language'])
        responses['Poster'].append(movie_data['Poster'])
        responses['Metascore'].append(movie_data['Metascore'])
        responses['imdbRating'].append(movie_data['imdbRating'])
        responses['BoxOffice'].append(movie_data['BoxOffice'])
        responses['Production'].append(movie_data['Production'])
        print('Good Data')

    except:
        print("Bad Data")
    pass    

    #delay for 65 seconds after every 100 calls    
    if i == 900:
        time.sleep(65)
        i = 0

#track number of call
        j += 1
    i += 1

else:
    print('------------------------------\nData Retrieval Complete\n------------------------------')
```

```python
#create dataframe from dictionary
movie_list_df_06_19 = pd.DataFrame.from_dict(responses, orient='index')
movie_list_df_06_19 = movie_list_df_06_19.transpose()

#export to csv
movie_list_df_06_19.to_csv ('output/movie_list_df_06_19.csv', header=True)

#display preview
movie_list_df_06_19
```
