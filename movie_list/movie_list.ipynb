```python
# Dependencies
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = 'https://www.the-numbers.com/box-office-records/domestic/all-movies/cumulative/released-in-2006'
```

```python
# Create a Beautiful Soup object
response = requests.get(url)
soup = bs(response.text, 'html.parser')
```

```python
# Extract all elements
#soup.body.find_all('b')
```

```python
# Print all movies
tds = soup.find_all('td')
# A blank list to hold the headlines
headlines = []
# Loop over td elements
for td in tds:
    # If td element has an anchor...
    if (td.b):
        # And the anchor has non-blank text...
        if (td.b.text):
            # Append the td to the list
            headlines.append(td)
```

```python
#get movie list
movies = []
for x in range(100):
    movies.append(headlines[x].text)
movies
```

```python
# Dependencies
import requests
from config import api_key

url = "http://www.omdbapi.com/?apikey=" + api_key + "&t="

responses1 = []

for movie in movies:
    movie_data = requests.get(url + movie).json()
    responses1.append(movie_data)
```
