```python
import pandas as pd
import datetime as dt
from re import sub
from decimal import Decimal
```

```python
load_df = pd.read_csv('output/upcoming_2021_2022.csv')
load_df = load_df.drop(['Production','Year','Plot','BoxOffice'], axis=1)
```

```python
load_df
```

```python
load_df = load_df.dropna(how='any')
load_df
```

```python
#converting to proper types
load_df['Released'] = pd.to_datetime(load_df['Released'], dayfirst=True)
load_df.Title = load_df.Title.astype('string')
load_df.Rated = load_df.Rated.astype('string')
load_df.Runtime = load_df.Runtime.astype('string')
load_df[['Runtime', 'Trash']] = load_df['Runtime'].str.split(' ', 1, expand=True)
load_df.Genre = load_df.Genre.astype('string')
load_df.Director = load_df.Director.astype('string')
load_df.Writer = load_df.Writer.astype('string')
load_df.Actors = load_df.Actors.astype('string')
load_df.Director = load_df.Director.astype('string')
load_df.Language = load_df.Language.astype('string')
load_df.Poster = load_df.Poster.astype('string')
#load_df['BoxOffice'] = load_df['BoxOffice'].replace('[\$,]', '', regex=True).astype(int)
load_df.dtypes
```

```python
 converted_df = load_df
```

```python
#clean dataframe filted before 2006
converted_df = pd.DataFrame(converted_df.rename(columns={"Unnamed: 0":"movie_id", 'Title':'title','Rated':'movie_rating','Released':'release_date',
'Runtime':'runtime', 'Genre': 'genre_name', 'Metascore':'metascore','imdbRating':'imdb_rating', 'BoxOffice': 'boxoffice'}))
#converted_df = converted_df.drop('Trash', axis=1)
converted_df = converted_df[~(converted_df['release_date'] < '2006-01-01')]

#export to csv
converted_df.to_csv ('output/clean_upcoming_df.csv', header=True, index=False)
converted_df
```

```python
#create movie table setup
movie = pd.DataFrame(converted_df[['movie_id','title','movie_rating', 'release_date', 'runtime', 'metascore', 'imdb_rating']])

#export to csv
movie.to_csv ('output/upcoming_movie.csv', header=True, index=False)

movie
```

```python
#create movie table setup
genre = pd.DataFrame(converted_df[['movie_id','genre_name']]).rename(columns={'movie_id':'genre_id'})
#export to csv
genre.to_csv ('output/upcoming_genre.csv', header=True, index=False)

genre.head()
```

```python
#create movie table setup
movie_genre_juction = movie.join(genre, how='outer')
movie_genre_juction = pd.DataFrame(movie_genre_juction[['movie_id','genre_id']])

#export to csv
genre.to_csv ('output/upcoming_movie_genre_juction.csv', header=True, index=False)

movie_genre_juction
```
