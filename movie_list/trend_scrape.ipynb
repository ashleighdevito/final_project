```python
from pytrends.request import TrendReq
import pandas as pd
import datetime as dt
from dateutil.relativedelta import relativedelta
```

```python
pytrend = TrendReq(hl='en-US', tz=360)
```

```python
# Import movie titles and release dates
movie_df = pd.read_csv("output/complete_movie_list_df.csv")
movie_df
```

```python
# fix datetime conversion
problem_entries = []
for index, row in movie_df.iterrows():
    try:
        title = row["Title"].split(': ')[0] + " movie"
        release_date = dt.datetime.strptime(row["Released"], "%d %b %Y")
        start_date = release_date - relativedelta(years=1)
        release_date = release_date.strftime("%Y-%m-%d")
        start_date = start_date.strftime("%Y-%m-%d")
        timeframe = f'{start_date} {release_date}'  
    except TypeError:
        print("bad data", title)
        problem_entries.append(title)
        pass
    # print(title, start_date, release_date)
    pytrend.build_payload(kw_list=[title], timeframe=timeframe')
    df = pytrend.interest_over_time()
    try:
        buzz = len(df.loc[df[title] >= 20])
    except KeyError:
        problem_entries.append(title)
        pass
    movie_df.loc[index, "Buzz"] = buzz
    print(title)
```
